```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
import pandas as pd
import numpy as np
import mglearn

iris = load_iris()

X_train, X_test, y_train, y_test = train_test_split(iris['data'], iris['target'], random_state=0, test_size=0.25)
```

```python
iris_df = pd.DataFrame(X_train, columns=iris.feature_names)
pd.plotting.scatter_matrix(iris_df, c=y_train, figsize=(15,15), marker='.', hist_kwds={'bins':20}, s=60, alpha=.8, cmap=mglearn.cm3)
```

```python
lr = RandomForestRegressor(n_estimators=5, random_state=2)
lr.fit(X_train, y_train)
pred = lr.predict(X_test)
pred
```

```python
df = {'실제값':y_test, '예측값':pred, '오차':y_test-pred}
df = pd.DataFrame(df)
df
```

```python
lr.estimators_
```

```python
for model in lr.estimators_:
    print("학습용 데이터 셋 평가 : {:.4f}".format(model.score(X_train,y_train)))
    print("테스트 데이터 셋 평가 : {:.4f}".format(model.score(X_test,y_test)))
    print()
```
